```python
# separating dependent and independent variable
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
#Data StandarliZation
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
#split the data into training and testing sets keeping 20% of the data for testing
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=9)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Decision tree
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.tree import DecisionTreeClassifier

model1=DecisionTreeClassifier(random_state=14)
dt=model1.fit(X_train,Y_train)
#prediction
Y_pred1=model1.predict(X_test)
#evaluatoion
accuracy_DT=accuracy_score(Y_test,Y_pred1)
f1_DT=f1_score(Y_test,Y_pred1)
precision_DT=precision_score(Y_test,Y_pred1)
recall_DT=recall_score(Y_test,Y_pred1)


#Random forest classifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.ensemble import RandomForestClassifier

model2=RandomForestClassifier(criterion='entropy', n_estimators=100)
rf=model2.fit(X_train,Y_train)
#prediction
Y_pred2=model2.predict(X_test)
#evaluatoion
accuracy_RF=accuracy_score(Y_test,Y_pred2)
f1_RF=f1_score(Y_test,Y_pred2)
precision_RF=precision_score(Y_test,Y_pred2)
recall_RF=recall_score(Y_test,Y_pred2)

#Support Vector Machine

from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.svm import SVC

model3=SVC(cache_size=100)
svm=model3.fit(X_train,Y_train)
#prediction
Y_pred3=model3.predict(X_test)
#evaluatoion
accuracy_SV=accuracy_score(Y_test,Y_pred3)
f1_SVF=f1_score(Y_test,Y_pred3)
precision_SV=precision_score(Y_test,Y_pred3)
recall_SV=recall_score(Y_test,Y_pred3)

#logistic regression
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.linear_model import LogisticRegression

model4= LogisticRegression(C=0.4,max_iter=1000,solver='liblinear')
lr=model4.fit(X_train,Y_train)
#prediction
Y_pred4=model4.predict(X_test)
#evaluatoion
accuracy_LR=accuracy_score(Y_test,Y_pred4)
f1_LR=f1_score(Y_test,Y_pred4)
precision_LR=precision_score(Y_test,Y_pred4)
recall_LR=recall_score(Y_test,Y_pred4)


#Nave Bayes
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.naive_bayes import  GaussianNB

gnb=GaussianNB()
gnb=gnb.fit(X_train,Y_train)
#prediction
gnb_pred=gnb.predict(X_test)
#evaluatoion
accuracy_GNB=accuracy_score(Y_test,gnb_pred)
f1_GNB=f1_score(Y_test,gnb_pred)
precision_GNB=precision_score(Y_test,gnb_pred)
recall_GNB=recall_score(Y_test,gnb_pred)



    
    
list1=['Decision Tree',' Random Forest Classifier','SVM','Logistic Regression','gnb']

list2=[accuracy_DT,accuracy_RF,accuracy_SV,accuracy_LR,accuracy_GNB]
list3=[f1_DT,f1_RF,f1_SVF,f1_LR,f1_GNB]
list4=[precision_DT,precision_RF,precision_SV,precision_LR,precision_GNB]
list5=[recall_DT,recall_RF,recall_SV,recall_LR,recall_GNB]

#df_Accuracy=pd.DataFrame({'Method Used':list1,'Accuracy':list2})
#print(df_Accuracy)

df_Accuracy=pd.DataFrame({'Method Used':list1,'Accuracy':list2,'f1score':list3,'Precison':list3,'recall':list4})
print(df_Accuracy)

chart=sns.barplot(x='Method Used',y='Accuracy',data=df_Accuracy)
chart.set_xticklabels(chart.get_xticklabels(),rotation=90)
print(chart)
```
